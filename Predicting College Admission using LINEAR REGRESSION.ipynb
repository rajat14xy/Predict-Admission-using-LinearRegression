# Calculating r^2 and Adjusted r^2

```python
def adjusted_rsquare(x,y):
    r2 = lr.score(x,y)
    n = x.shape[0]   # NUMBER OF DATA
    p = x.shape[1]   # NUMBER OF FEATURES
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

```python
print({'Accuracy score':lr.score(X_train,y_train)})
print({'Adjusted r^2':adjusted_rsquare(X_train,y_train)})
```

```python
print({'Accuracy score':lr.score(X_test,y_test)})
print({'Adjusted r^2':adjusted_rsquare(X_test,y_test)})
```

# Checking OVERFIT using Regularization

**Using Elasticnet so that if possible some of the unwanted features will die out**

```python
from sklearn.linear_model  import Ridge,Lasso,RidgeCV, LassoCV, ElasticNet, ElasticNetCV, LinearRegression
```

```python
elasticnet = ElasticNetCV(alphas=None,cv=4)
elasticnet.fit(X_train,y_train)
```

```python
elasticnet.alpha_
```

```python
elasticnet.l1_ratio
```

```python
elasticnet_regularization = ElasticNet(alpha=elasticnet.alpha_,l1_ratio=0.5)
elasticnet_regularization.fit(X_train,y_train)
```

```python
print({'r^2 Score Train':elasticnet_regularization.score(X_train,y_train)})
print({'r^2 Score Test':elasticnet_regularization.score(X_test,y_test)})
```

**We can see our values of r^2 are close. Therefore we can say our model is not not OVERFIT.**
